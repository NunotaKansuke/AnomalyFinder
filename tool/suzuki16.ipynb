```python
import numpy as np
import matplotlib.pyplot as plt
import AnomalyFinder as af
```

```python
list_path = "../suzuki16/list"
names = np.genfromtxt(list_path,dtype=str)
```

```python
tmp= af.AnomalyFinder()

file_name = f"../suzuki16/data/MOAR_{names[0]}.dat"
print(names[0])
tmp.set_data(file_name,subtract_2450000=False)

t_range=[4200,4300]
tmp.plot_data(markersize=5,t_range=[t_range[0],t_range[1]])
plt.show()

tmp.PSPL_fit(4245,10,0.01)
print(tmp.t0_PSPL,tmp.tE_PSPL,tmp.u0_PSPL)
```

```python
tmp.plot_residual_from_PSPL(t_range=[t_range[0],t_range[1]],markersize=3,capsize=2)

plt.subplots_adjust(left=0.15, right=0.95, top=0.95, bottom=0.08)

plt.show()
```

```python
tmp.run_grid_search(t_range=[t_range[0],t_range[1]],teff_init=0.3,teff_grid=13,common_ratio=4/3)
```

```python
tmp.show_grid_search_result(s=1)

plt.show()
```

```python
tmp.show_anomaly_signal(which="zero",markersize=3,capsize=2)

tmp.show_anomaly_signal(which="flat",markersize=3,capsize=2)
```
