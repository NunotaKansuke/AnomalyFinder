```python
import numpy as np
import matplotlib.pyplot as plt
import AnomalyFinder as af
```

```python
tmp= af.AnomalyFinder()
tmp.set_data("../lcurve/MOA-2010-BLG-073.dat",subtract_2450000=True)
tmp.cut_data(t_range=[4900,5700])

tmp.plot_data(markersize=5,t_range=[5230,5430])
plt.show()


tmp.PSPL_fit(5320,600,0.01)
print(tmp.t0_PSPL,tmp.tE_PSPL,tmp.u0_PSPL)
```

```python
tmp.plot_residual_from_PSPL(t_range=[5230,5430],markersize=3,capsize=2)

plt.subplots_adjust(left=0.2, right=0.88, top=0.95, bottom=0.08)

plt.savefig("../image/MB1073_res.png")
plt.show()
```

```python
tmp.run_grid_search(t_range=[5330,5335],teff_init=0.01,teff_grid=25)
```

```python
# rcParams["font.size"] = 13
tmp.show_grid_search_result(s=1)

# plt.subplots_adjust(left=0.075, right=0.88, top=0.95, bottom=0.08)

plt.savefig("../image/MB1073_grid.png")
```

```python
# plt.figure(figsize=(6,2))
tmp.show_anomaly_signal(which="zero",markersize=3,capsize=2)
# plt.subplots_adjust(left=0.2, right=0.99, top=0.9, bottom=0.21)
plt.savefig("../image/MB1073_zero.png")

# plt.figure(figsize=(6,2))
tmp.show_anomaly_signal(which="flat",markersize=3,capsize=2)
# plt.subplots_adjust(left=0.2, right=0.99, top=0.9, bottom=0.21)
plt.savefig("../image/MB1073_flat.png")
```
