```python
import numpy as np
import matplotlib.pyplot as plt
import AnomalyFinder as af
```

```python
tmp= af.AnomalyFinder()
tmp.set_data("../lcurve/good_example_01.dat",subtract_2450000=False)
t_range=[4000,5500]

tmp.plot_data(markersize=5,t_range=[t_range[0],t_range[1]])
plt.show()

tmp.PSPL_fit(4625,30,0.01)
print(tmp.t0_PSPL,tmp.tE_PSPL,tmp.u0_PSPL)
```

```python
tmp.plot_residual_from_PSPL(t_range=[t_range[0],t_range[1]],markersize=3,capsize=2)

plt.subplots_adjust(left=0.15, right=0.95, top=0.95, bottom=0.08)

plt.savefig("../image/good_example_01_res.png")
plt.show()
```

```python
tmp.run_grid_search(t_range=[t_range[0],t_range[1]],teff_init=0.3,teff_grid=13,common_ratio=4/3)
```

```python
# rcParams["font.size"] = 13
tmp.show_grid_search_result(s=1)

# plt.subplots_adjust(left=0.075, right=0.88, top=0.95, bottom=0.08)

plt.savefig("../image/good_example_01_grid.png")
```

```python
# plt.figure(figsize=(6,2))
tmp.show_anomaly_signal(which="zero",markersize=3,capsize=2)
# plt.subplots_adjust(left=0.2, right=0.99, top=0.9, bottom=0.21)
plt.savefig("../image/good_example_01_zero.png")

# plt.figure(figsize=(6,2))
tmp.show_anomaly_signal(which="flat",markersize=3,capsize=2)
# plt.subplots_adjust(left=0.2, right=0.99, top=0.9, bottom=0.21)
plt.savefig("../image/good_example_01_flat.png")
```
